```python
import pandas as pd
```

```python
df = pd.read_csv(r'data\LLLG.csv')
df.head()
```

```python
df = df.loc[:, ~df.columns.str.contains('Voltages', case=False)]
df.head()
```

```python
df['Fault'] = df['Time'].apply(lambda t: 'Yes' if 0.2 <= t <= 0.4 else 'No') 
df.head()
```

```python
df['Absolute Ia'] = df.filter(like='MAINIA').abs().iloc[:, 0]
df['Absolute Ib'] = df.filter(like='MAINIB').abs().iloc[:, 0]
df['Absolute Ic'] = df.filter(like='MAINIC').abs().iloc[:, 0]
```

```python
df = df[['Time', 'Absolute Ia', 'Absolute Ib', 'Absolute Ic', 'Fault']]
df.head()
```

```python
df.to_csv('data_modified/LLLG_modified.csv', index=False)
```
